## Oxford-IIIT Pet Image Classification using CNN & Transfer Learning.

This project implements a Convolutional Neural Network (CNN) model to classify pet images into different breeds using the Oxford-IIIT Pet Dataset. The goal is to build an end-to-end deep learning pipeline for image classification, including data preprocessing, model training, evaluation, and model saving.

# Model 1: CNN

```python
import sys
print(sys.executable)
```

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
```

## Load the Dataset (Oxford-IIIT Pet)

```python
(ds_train_raw, ds_test_raw), ds_info = tfds.load(
    "oxford_iiit_pet",
    split=["train", "test"],
    with_info=True,
    as_supervised=True,  # ÙŠØ±Ø¬Ù‘Ø¹ (image, label) Ù…Ø¨Ø§Ø´Ø±Ø©
)

num_classes = ds_info.features["label"].num_classes
print("Number of classes:", num_classes)
print("Train examples:", ds_info.splits["train"].num_examples)
print("Test examples:", ds_info.splits["test"].num_examples)
```

## Splitting Training Data into Train and Validation Sets

```python
train_size = ds_info.splits["train"].num_examples
val_ratio = 0.2
val_size = int(train_size * val_ratio)

ds_val_raw = ds_train_raw.take(val_size)
ds_train_raw = ds_train_raw.skip(val_size)

print("Approx train size:", train_size - val_size)
print("Approx val size:", val_size)
```

## Visualizing Sample Images from the Training Set

```python
import matplotlib.pyplot as plt
def show_sample(ds, n=9):
    plt.figure(figsize=(8, 8))
    for i, (image, label) in enumerate(ds.take(n)):
        plt.subplot(3, 3, i + 1)
        plt.imshow(image)
        plt.axis("off")
        plt.title(f"Label: {label.numpy()}")
    plt.tight_layout()
    plt.show()

show_sample(ds_train_raw)
```

## Data Preprocessing and Augmentation

```python
IMG_SIZE = 160  
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

def preprocess_image(image, label):
    image = tf.cast(image, tf.float32)
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image / 255.0
    return image, label

def augment_image(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, max_delta=0.1)
    image = tf.image.random_contrast(image, lower=0.9, upper=1.1)
    return image, label
```

## Data Pipeline Implementation

```python
ds_train = ds_train_raw.map(preprocess_image, num_parallel_calls=AUTOTUNE)
ds_train = ds_train.map(augment_image, num_parallel_calls=AUTOTUNE)
ds_train = ds_train.shuffle(1000)
ds_train = ds_train.batch(BATCH_SIZE)
ds_train = ds_train.prefetch(AUTOTUNE)
```

Validation Dataset

```python
ds_val = ds_val_raw.map(preprocess_image, num_parallel_calls=AUTOTUNE)
ds_val = ds_val.batch(BATCH_SIZE)
ds_val = ds_val.prefetch(AUTOTUNE)
```

Test Dataset

```python
ds_test = ds_test_raw.map(preprocess_image, num_parallel_calls=AUTOTUNE)
ds_test = ds_test.batch(BATCH_SIZE)
ds_test = ds_test.prefetch(AUTOTUNE)
```

## Model 1: Convolutional Neural Network (CNN)

```python
from tensorflow.keras import layers, models

def build_better_cnn(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=num_classes):
    inputs = layers.Input(shape=input_shape)

    x = layers.Conv2D(32, (3, 3), padding="same", activation="relu")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2))(x)

    x = layers.Conv2D(64, (3, 3), padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2))(x)

    x = layers.Conv2D(128, (3, 3), padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2))(x)

    x = layers.Conv2D(256, (3, 3), padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2))(x)

    # Ø¨Ø¯Ù„ Flatten
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    return model

model = build_better_cnn()
model.summary()
```

## Training Setup (Optimizer, Loss Function, Metrics)

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
) 
```

## The model is trained using EarlyStopping to prevent overfitting and ModelCheckpoint to save the best model based on validation accuracy.


```python
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    "best_pets_cnn.h5",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max",
    verbose=1
)

earlystop_cb = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True,
    verbose=1
)

history = model.fit(
    ds_train,
    validation_data=ds_val,
    epochs=30,
    callbacks=[checkpoint_cb, earlystop_cb]
)
```

## Training Performance Visualization

```python
def plot_history(history):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 4))

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label="Train Acc")
    plt.plot(epochs_range, val_acc, label="Val Acc")
    plt.legend()
    plt.title("Accuracy")

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label="Train Loss")
    plt.plot(epochs_range, val_loss, label="Val Loss")
    plt.legend()
    plt.title("Loss")

    plt.show()

plot_history(history)
```

## Final Evaluation on Test Set

```python
test_loss, test_acc = model.evaluate(ds_test)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

## Sample Predictions Visualization

```python
class_names = ds_info.features["label"].names
print("Number of classes:", len(class_names))
print("Some classes:", class_names[:10])




images_batch, labels_batch = next(iter(ds_test))

pred_probs = model.predict(images_batch)
pred_labels = tf.argmax(pred_probs, axis=1)

plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images_batch[i])
    true_label = class_names[labels_batch[i].numpy()]
    pred_label = class_names[pred_labels[i].numpy()]
    plt.title(f"T: {true_label}\nP: {pred_label}", fontsize=8)
    plt.axis("off")
plt.tight_layout()
plt.show()


```

## Model Performance Summary

The model shows clear signs of overfitting, as the training accuracy reaches 63% while the validation accuracy remains around 14%. 

The CNN trained from scratch struggled to generalize due to the limited dataset size and the complexity of classifying 37 different breeds. As a result, it achieved high training accuracy but very low validation and test performance.

The performance of the CNN could potentially be improved by adding more convolutional layers or increasing model complexity. Such modifications would require more computational resources and longer training time.

To address this issue, transfer learning with MobileNetV2 was applied. Since the model was pre-trained on a large-scale dataset, it provided strong feature extraction and significantly improved generalization compared to training from scratch.


# Model 2: Transfer Learning (MobileNetV2)

## 1-Imports the necessary Libs

```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras import layers, models
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
```

## 2-Global config

```python
IMG_SIZE = 160     
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE
```

## 3- Load Oxford-IIIT Pet dataset

```python
(ds_train_raw, ds_test_raw), ds_info = tfds.load(
    "oxford_iiit_pet",
    split=["train", "test"],
    with_info=True,
    as_supervised=True  # (image, label)
)

num_classes = ds_info.features["label"].num_classes
print("Number of classes:", num_classes)
```

## 4-Create train / validation split from train

```python
train_size = ds_info.splits["train"].num_examples
val_size = int(0.2 * train_size)

ds_val_raw = ds_train_raw.take(val_size)
ds_train_raw = ds_train_raw.skip(val_size)

print("Train examples:", train_size - val_size)
print("Val   examples:", val_size)
print("Test  examples:", ds_info.splits["test"].num_examples)
```

## 5- Data Preprocessing and Augmentation

```python

def preprocess_image(image, label):
    """
    - Resize to fixed size
    - Convert to float32
    - Apply MobileNetV2 preprocess_input (scales to [-1,1])
    """
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = tf.cast(image, tf.float32)      
    image = preprocess_input(image)      
    return image, label

def augment_image(image, label):
    """
    Simple augmentation:
    - Random horizontal flip
    - Random brightness / contrast
    """
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, max_delta=0.1)
    image = tf.image.random_contrast(image, lower=0.9, upper=1.1)
    image = tf.clip_by_value(image, -1.0, 1.0)
    return image, label

def prepare_dataset(ds, training=False):
    """
    - Apply preprocessing
    - (Optional) augmentation
    - Shuffle (if training)
    - Batch + prefetch
    """
    ds = ds.map(preprocess_image, num_parallel_calls=AUTOTUNE)
    
    if training:
        ds = ds.map(augment_image, num_parallel_calls=AUTOTUNE)
        ds = ds.shuffle(1000)
    
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(AUTOTUNE)
    return ds

train_ds = prepare_dataset(ds_train_raw, training=True)
val_ds   = prepare_dataset(ds_val_raw,   training=False)
test_ds  = prepare_dataset(ds_test_raw,  training=False)
```

## 6- Build Transfer Learning Model

```python
def build_transfer_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=num_classes):
    # 1) Base model (MobileNetV2 pretrained on ImageNet)
    base_model = MobileNetV2(
        input_shape=input_shape,
        include_top=False,     # remove original classifier
        weights="imagenet"
    )
    
    # Freeze base model in first phase
    base_model.trainable = False

    inputs = layers.Input(shape=input_shape)
    x = base_model(inputs, training=False) 

    # Global average pooling  
    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dropout(0.3)(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs, name="pet_mobilenetv2_classifier")
    return model

model = build_transfer_model()
model.summary()
```

## 7-Training Setup (Optimizer, Loss Function, Metrics)

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=["accuracy"]
)
```

## 8- Callbacks

```python
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    "best_pet_mobilenetv2.h5",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max"
)

earlystop_cb = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    restore_best_weights=True
)
```

## 9- Train

```python
EPOCHS_PHASE1 = 12  

history_phase1 = model.fit(
    train_ds,
    epochs=EPOCHS_PHASE1,
    validation_data=val_ds,
    callbacks=[checkpoint_cb, earlystop_cb]
)
```

## 10- Evaluate

```python
print("\n=== Evaluation after Phase 1 (frozen base) ===")
test_loss, test_acc = model.evaluate(test_ds)
print(f"Test accuracy: {test_acc:.4f}")
```

## 11- Training Performance Visualization

```python
import matplotlib.pyplot as plt
def plot_history(history):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 4))

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label="Train Acc")
    plt.plot(epochs_range, val_acc, label="Val Acc")
    plt.legend()
    plt.title("Accuracy")

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label="Train Loss")
    plt.plot(epochs_range, val_loss, label="Val Loss")
    plt.legend()
    plt.title("Loss")

    plt.show()
plot_history(history_phase1)
```

## 13 -Fine-tuning (Optional )

```python
# Load best weights from Phase 1
model.load_weights("best_pet_mobilenetv2.keras")

# Unfreeze the base model
base_model.trainable = True

# Freeze first layers, fine-tune only top layers
fine_tune_at = 100

for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Recompile with lower learning rate
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

# Train again
history_phase2 = model.fit(
    train_ds,
    epochs=8,
    validation_data=val_ds,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=3,
            restore_best_weights=True
        )
    ]
)

# Final evaluation
test_loss, test_acc = model.evaluate(test_ds)
print(f"Final Test accuracy after Fine-Tuning: {test_acc:.4f}")
plot_history(history_phase2)
```
